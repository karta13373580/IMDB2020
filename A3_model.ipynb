```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
get_ipython().run_line_magic('matplotlib', 'inline')
from sklearn import preprocessing
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, Activation, Dropout
from sklearn.preprocessing import MinMaxScaler
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn import metrics
```

```python
data_train=pd.read_csv("A3_train.csv")
data_test=pd.read_csv("y_train.csv")
```

```python
min_max = MinMaxScaler()
data_train = pd.DataFrame(min_max.fit_transform(data_train), 
                                     columns=data_train.keys())
```

```python
X,y=shuffle(data_train,data_test,random_state=2)
X_train,X_test,y_train,y_test=train_test_split(data_train,data_test['Input_A3_013'],test_size=0.2,random_state=2)
```

```python
model = Sequential()
model.add(Dense(640, activation='relu', input_dim=104))
model.add(Dense(400, activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(300, activation='relu'))
model.add(Dense(200, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(units=10, activation='softmax'))
model.add(Dropout(0.1))
model.add(Dense(1))


# model = Sequential()
# model.add(Dense(384, activation='relu', input_dim=105))
# model.add(Dense(150, activation='relu'))
# model.add(Dense(64, activation='relu'))
# model.add(Dense(units=10, activation='softmax'))
# model.add(Dropout(0.1))
# model.add(Dense(1))
```

```python
opt = keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss='mse', optimizer=opt, metrics=['acc'])
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=300, batch_size=30, verbose=1)
```

```python
#
min_max = MinMaxScaler()
X_test = pd.DataFrame(min_max.fit_transform(X_test), 
                                     columns=X_test.keys())

v_prediction=model.predict(X_test)

print(np.sqrt(metrics.mean_squared_error(y_test, v_prediction)))
plt.plot(v_prediction)
```

```python
#
test=pd.read_csv("A3_test.csv")
min_max = MinMaxScaler()
test = pd.DataFrame(min_max.fit_transform(test), 
                                     columns=test.keys())

A3_13_prediction=model.predict(test)
plt.plot(A3_13_prediction)
```

```python
A3_13_prediction = pd.DataFrame(A3_13_prediction)
result=pd.read_excel("result.xlsx")
result['Input_A3_013']=A3_13_prediction
```
